```python
fdict={
'Urea':1,
'DAP':2,
'14-35-14':3,
'28-28':4,
'17-17-17':5,
'20-20':6,
'10-26-26':7,
}
```

```python
df['fert_no']=df['Fertilizer Name'].map(fdict)
```

```python
df['fert_no'].value_counts()
```

```python
df.drop('Fertilizer Name',axis=1,inplace=True)
df.head()
```

```python
df.select_dtypes(include=['object']).columns
```

```python
lb = LabelEncoder()
df["Soil Type"]=lb.fit_transform(df['Soil Type'])
df['Crop Type']=lb.fit_transform(df['Crop Type'])
```

```python
df.head()
```

```python
x=df.drop('fert_no',axis=1)
y=df['fert_no']
print(f"The shape of features is: {x.shape}")
print(f"The shape of target is: {y.shape}")
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, train_size=0.8, random_state=42)

print(f"The shape of x_train is: {x_train.shape}")
print(f"The shape of x_test is: {x_test.shape}")
print(f"The shape of y_train is: {y_train.shape}")
print(f"The shape of y_test is: {y_test.shape}")
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
test_accuracy=model.score(x_test, y_test)
print(f"The accuracy of the model is: {test_accuracy*100:.2f}%")
```

```python
train_accuracy=model.score(x_train, y_train)
print(f"The accuracy of the model on the training set is: {train_accuracy*100:.2f}%")
```

```python
def recommend_fertilizer(Temparature, Humidity, Moisture, Soil_Type, Crop_Type, Nitrogen, Potassium, Phosphorous):
    features = np.array([[Temparature, Humidity, Moisture, Soil_Type, Crop_Type, Nitrogen, Potassium, Phosphorous]])
    transformed_features = sc.transform(features)
    prediction = model.predict(transformed_features).reshape(1,-1)
    fert_dict = {1: 'Urea', 2: 'DAP', 3: '14-35-14', 4: '28-28', 5: '17-17-17', 6: '20-20', 7: '10-26-26'}
    fertilizer = [fert_dict[i] for i in prediction[0]]
    
    return f"{fertilizer} is a best fertilizer for the given conditions" 
```

```python
t,h,m,s,c,n,p,t=56,0.5,0.6,2,3,10,15,6
recommend_fertilizer(t,h,m,s,c,n,p,t)
```
